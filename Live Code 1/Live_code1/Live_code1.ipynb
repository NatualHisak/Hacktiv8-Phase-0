a. Buatlah dua fungsi eksplisit masing-masing memuat perhitungan hambatan (R) dan daya (P) yang nantinya akan digunakan untuk perhitungan

Import library yang digunakan yaitu pandas

load dataset dengan pd.read_excel("path") karena dataset berupa file excel

```python
df = pd.read_excel(r'C:\Users\Omen\Desktop\batch_16\Live Code 1\Live_code1\NMC1-9.xlsx') #Load Dataset
df #tampilkan tabel
```

buat fungsi perhitungan nilai R dan P dengan menggunakan def () :

```python
def R(V,I) : #fungsi menghitung nilai R dengan input berupa nilai V dan I
    R = V/I #rumus hitung R
    return R #return nilai R 

def P(V,I) : #fungsi menghitung nilai P dengan input berupa nilai V dan I
    P = V*I #rumus hitung P
    return P #return nilai P
```

b. Buat kolom baru dengan nama kolom R/P berisikan hasil perhitungan daya/hambatan dengan kriteria untuk Kapasitas >=2.5 hitung daya, Kapasitas <2.5 hitung nilai hambatan. (Hint: Gunakan for loop dan if statement untuk mengerjakan perintah nomor b, kamu bisa menggunakan list untuk menyimpan perhitungan sementara, setelahnya dapat dimasukkan ke kolom R/P).

memasukkan nilai pada kolom V_m,Current (A), dan Cap kedalam list dan mendefine list kosong untuk hasil perhitungan

```python
RP = [] #define list kosong dalam variabel RP sebagai tempat menampung value hasil hitung R atau P
V = df['V_m'].tolist() # ubah kolom V_m menjadi list
I = df['Current (A)'].tolist() # ubah kolom Current (A) menjadi list
C = df['Cap'].tolist()# ubah kolom Cap menjadi list
```

Memasukkan nilai hasil hitung R atau P kedalam list kosong dengan memanfaatkan loop dan conditional serta fungsi yang telah dibuat pada soal a

```python
y=0 
z=0
for x in C : #For loop untuk menjalankan program tiap nilai pada kolom Cap
    if x>=2.5 : #conditional untuk nilai kapasitas lebih dari sama dengan 2.5
        RP.append(P(V[y],I[z])) #hitung nilai P dan hasilnya dimasukkan ke list RP
    elif x<2.5 :#conditional untuk nilai kapasitas kurang dari 2.5
        RP.append(R(V[y],I[z])) #hitung nilai R dan hasilnya dimasukkanke list RP
    y=y+1
    z=z+1
RP # tampilkan list RP yang sudah lengkap dengan value nya
```

Copy dataframe df ke dalam variabel baru

```python
new_df = df.copy() #copy df kedalam variabel new_df
new_df #tampilkan tabel hasil copy yang ada pada variabel new_df
```

Tambahkan kolom baru berdasarkan list RP yang telah dibuat

```python
new_df["R/P"]=RP #buat kolom baru bernama R/P dengan isi berupa value pada list RP
new_df #tampilkan tabel setelah ditambahkan kolom baru
```
