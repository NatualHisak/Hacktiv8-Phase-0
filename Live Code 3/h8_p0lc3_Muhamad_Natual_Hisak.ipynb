7. Berapa lama rata-rata waktu (dalam hari) kasus tindak kejahatan 'theft' terselesaikan sejak kasus tersebut terangkat selama tahun 2016? (Hint: Hitung selisih hari dari timestamp hingga clearance date)

Query : 

```
SELECT AVG(days) as average_days 
FROM(
    SELECT day_cleared,day_reported,
           DATE_DIFF(day_cleared, day_reported,day) as days 
    FROM (
      SELECT
        timestamp as day_reported,
        clearance_date as day_cleared
      FROM `bigquery-public-data.austin_crime.crime` 
      where timestamp >= "2016-01-01" and timestamp <= "2016-12-31" and primary_type = "Theft"
      GROUP BY 1,2))

```
