<h1>NON GRADED CHALLENGE 8</h1>

Carilah gradient dari sebuah fungsi:
$$3x+4y=5$$

Import library

```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt 
```

Gradien X

fungsi diatas bisa ditulis menjadi 

$$ y = 1.25 - 0.75x$$

jadikan x dan y sebagai variabel dengan range -50 sampai 50

```python
x = np.linspace(-50, 50)
y = np.linspace(-50, 50)
```

Persamaan baru

```python
y2 = 1.25 - 0.75*x
y2
```

```python
grax = np.gradient(y2) #Mencari gradien x
grax
```

Plot

```python
fig, ax = plt.subplots(ncols=2, figsize=(20,5))

ax[0].plot(x, y2)
ax[0].set_title('x')

ax[1].plot(grax, y2)
ax[1].set_title('grax')
plt.xlim(-2, 2)
```

Gradien X

fungsi diatas bisa ditulis menjadi 

$$ x = 5/3 - 4/3y$$

```python
x2 = 5/3 - 4/3*y #persamaan baru
x2
```

```python
gray = np.gradient(x2) #gradient Y 
gray
```

Plot

```python
fig, ax = plt.subplots(ncols=2, figsize=(21,5))

ax[0].plot(x2, y)
ax[0].set_title('y')

ax[1].plot(x2, gray)
ax[1].set_title('gray')
plt.ylim(-3, 3)
```
